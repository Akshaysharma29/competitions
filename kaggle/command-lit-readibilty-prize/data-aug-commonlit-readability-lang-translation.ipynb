# pipeline

## Read data

```python
import numpy as np # linear algebra
import pandas as pd
data = pd.read_csv('../input/commonlitreadabilityprize/train.csv')
print(len(data))
data.head()
```

## French

```python
from transformers import MarianTokenizer, MarianMTModel
from typing import List
src = 'en'  # source language
trg = 'fr'  # target language
sample_text = "How are you"
model_name = f'Helsinki-NLP/opus-mt-{src}-{trg}'
model_en_fr = MarianMTModel.from_pretrained(model_name)
tokenizer_en_fr = MarianTokenizer.from_pretrained(model_name)
batch = tokenizer_en_fr([sample_text], return_tensors="pt")
gen = model_en_fr.generate(**batch)
tokenizer_en_fr.batch_decode(gen, skip_special_tokens=True)
```

```python
src = 'fr'  # source language
trg = 'en'  # target language
sample_text = "Comment allez-vous?"
model_name = f'Helsinki-NLP/opus-mt-{src}-{trg}'
model_fr_en = MarianMTModel.from_pretrained(model_name)
tokenizer_fr_en = MarianTokenizer.from_pretrained(model_name)
batch = tokenizer_fr_en([sample_text], return_tensors="pt")
gen = model_fr_en.generate(**batch)
tokenizer_fr_en.batch_decode(gen, skip_special_tokens=True)
```

```python
excerpt = []
target = []
for i ,row in data.iterrows():
    excerpt.append(row['excerpt'])
    target.append(row['target'])
```

```python
from tqdm import tqdm
fr_excerpt= []
fr_target = []
batch_size = 32
for i in tqdm(range(0, len(excerpt), batch_size)):
    if i == len(excerpt):
        break
    sample_text = excerpt[i:i+batch_size]
    batch = tokenizer_en_fr(sample_text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    gen = model_en_fr.generate(**batch)
    sample_text=tokenizer_en_fr.batch_decode(gen, skip_special_tokens=True)
    batch = tokenizer_fr_en(sample_text, return_tensors="pt",  padding=True, truncation=True, max_length=512)
    gen = model_fr_en.generate(**batch)
    fr_excerpt = fr_excerpt+tokenizer_fr_en.batch_decode(gen, skip_special_tokens=True)
    fr_target = fr_target+target[i:i+batch_size]
#     if i==4:
#         break
```
